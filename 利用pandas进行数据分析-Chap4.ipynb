### 布尔型索引

```python
names = np.array(['Bob','Joe','Will','Bob','Will','Joe','Joe'])
names == 'Bob' #数组的比较运算会生成一个布尔型数组
```

```python
data = np.random.randn(7,4) #生成随机函数
data
```

```python
#使用布尔数组进行数组索引
data[names=='Bob'] #获得x=0,3的部分
```

```python
data[[0,3]] #同上
```

```python
data[names=='Bob',:2]
```

```python
mask = (names == 'Bob') | (names == 'Will') # &-与，|-或, 布尔型数组中and 和 or 无效
data[mask]
```

### 利用数组进行数据处理

```python
points = np.arange(-5,5,0.01)
points
```

```python
xs,ys = np.meshgrid(points,points) #meshgrid生成网格型数据，接收两个一维数组，生成2个二维数组
```

```python
xs
```

```python
ys
```

```python
import matplotlib.pyplot as plt
z = np.sqrt(xs**2+ys**2)
z
```

```python
plt.imshow(z)
plt.colorbar()
plt.title('image plot of $\sqrt{x^2+y^2}$ for a grid of values')
plt.show()
```

```python
#使用数组运算表达条件逻辑
xarr = np.array([1.1,1.2,1.3,1.4,1.5])
yarr = np.array([2.1,2.2,2.3,2.4,2.5])
cond = np.array([True,False,True,True,False])
```

```python
result = [(x if c else y)
         for x,y,c in zip(xarr,yarr,cond)]
result
```

```python
result = np.where(cond,xarr,yarr) #适用于多维数组，且能够实现嵌套用于复杂逻辑
result
```
